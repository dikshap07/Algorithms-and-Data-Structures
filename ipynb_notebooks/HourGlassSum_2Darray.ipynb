```python
arr = []
for _ in range(6):
        arr.append(list(map(int, input().rstrip().split())))
```

```python
def hourglassSum(arr):
    
    hourglassSums = []
    for j in range(0,4):
      for i in range(0,4):
        print(f'i {i} , j {j}')

        one_Sum = arr[j][i]+arr[j][i+1] + arr[j][i+2] + arr[j+1][i+1] + arr[j+2][i]+   arr[j+2][i+1] + arr[j+2][i+2]
        
        hourglassSums.append(one_Sum)
    print(hourglassSums)
    return max(hourglassSums)
```

```python
result = hourglassSum(arr)
result
```
