###Problem -01

Check if a key is present in every segment of size k in an array.Capture.PNG

```python
def check_element(A,element,segment_size):

  start= 0

  #print(len(A))

  while start < (len(A)-segment_size):  
    
    
    for i in range(start , (start+segment_size)):   
      
      if A[i] == element:   # if element present in segment move to next segment 
         
        start += segment_size
        break

      elif i == (start+segment_size) -1 and A[i] != element:    #if we have reached last element in the segment and it is still not equal to the required element than return 'No'

        return 'No' 
        
      else:   #else move to next element in the segment

        i+=1
      
  
  return 'Yes'   #return Yes only when all segments have been checked and they have the element because while loop will only complete if all checked and element present
```
