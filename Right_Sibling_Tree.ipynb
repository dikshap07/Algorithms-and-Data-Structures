##Problem: Given a Binary Tree, change the tree such that each nodes points to its immediate right neighbour on the same level. If a node doesnt have an immediate neighbour on the right it doesnt just take the first neighbour/sibling on the right, we leave it as it is.

Eg.:
image.png

```python
#O(N):Time  | N: No of nodes
#O(d) : Space | d:  depth of the tree


def right_sibling_tree(node):

  mutate(node,None,None)

  return node

def mutate(node,parent,isLeftChild):  #takes node, its parent, and isLeftChild: a booelean to check if out node is the child or not

  #base case: if we reach the end of the tree
  if node is None:
    return

  left_child,right_child = node.left,node.right   #we keep the children here so that when connection is severed btw node and its right child we dont loose what the right child was


  #step : 1  mutate function on the left child of the node
  
  mutate(left_child,node,True) #true because we are working on leftchild

  #step:2 transformation

  #case 1: when the parent of the node we are working on is None
  if parent is None:

    node.right = None
  
  #case 2: if it is left child of our parent Node
  elif isLeftChild:

    node.right = parent.right #we will make the right pointer of our node towards the right child of the parent node
                              #bcz our node is the left node so immidiate neighbour is its parent node's right child(or we can say its sibling)

  else:  #if our node is right child of parent node

    #check is its parent's right pointer is none or not(bcz now we have already changed the right pointer of parent node from right child to right sibling )
    if parent.right is None:
      node.right = None  #if the right pointer is still None then the right pointer for the node will also be none

    else:
      node.right = parent.right.left #if right pointer of parent is not None then our node will point to the left child of this right pointer of parent


  #step: 3 mutate function on the right child of the node

  mutate(right_child,node,False)

```

```python
class BinaryTree:

  def __init__(self,value):

    self.value = value
    self.left = None
    self.right = None
    self.parent = None



bt = BinaryTree(1)
bt.parent = None
bt.left = BinaryTree(2)
bt.right = BinaryTree(3)
bt.left.parent = bt
bt.right.parent = bt
bt.left.left = BinaryTree(4)

bt.left.left.parent = bt.left
k = bt.left.left
k.right = BinaryTree(8)
k.right.parent = k
bt.right.left = BinaryTree(6)
bt.right.right = BinaryTree(7)


l = bt.right 
l.right.parent = l
l.left.parent = l
```
